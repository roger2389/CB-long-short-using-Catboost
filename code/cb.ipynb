```python
import re

def find_active_cb_for_stock(stock_ticker, rebal_date, cb_ret, lookback_days=5):
    pattern = re.compile(rf'^{stock_ticker}(\d+)\s?.*$')

    possible_cbs = []
    for cb_col in cb_ret.columns:
        m = pattern.match(cb_col)
        if m:
            digits_str = m.group(1)
            digits_val = int(digits_str)
            possible_cbs.append((cb_col, digits_val))

    if not possible_cbs:
        return None

    if rebal_date not in cb_ret.index:
        return None

    idx_loc = cb_ret.index.get_loc(rebal_date)
    end_loc = idx_loc + (lookback_days - 1)
    if end_loc >= len(cb_ret.index):
        return None

    valid_cbs = []
    for cb_col, digits_val in possible_cbs:
        sub_series = cb_ret.iloc[idx_loc:end_loc+1][cb_col]
        if sub_series.notna().all() and not sub_series.eq(0).all():
            valid_cbs.append((cb_col, digits_val))

    if not valid_cbs:
        return None

    # 若要取後面數字最大的
    best = max(valid_cbs, key=lambda x: x[1])
    return best[0] 

def calculate_return_series(
    ret, Top_n, Low_N, AUM, adj_close, exp_ret,
    cb_ret=None, cb_adj_close=None, cb_exp_ret=None,
    stop_loss_threshold=-0.1,
    lookback_days=5,bench_exp_ret = None
):

    month_end_dates = ret.groupby(ret.index.to_period('W')).tail(1).index
    ret = ret.rolling(window=20, min_periods=1).mean().dropna()

    def fill_signal(signal, weight):
        filled_signal = pd.DataFrame(index=weight.index, columns=weight.columns)
        first_date = month_end_dates[0]
        filled_signal.loc[:first_date] = signal.loc[first_date].values
        for i in range(len(month_end_dates) - 1):
            current_date, next_date = month_end_dates[i], month_end_dates[i + 1]
            date_range = weight.loc[current_date:next_date].index
            filled_signal.loc[date_range] = signal.loc[current_date].values
        return filled_signal.dropna(how='all')

    ranking_desc = ret.rank(axis=1, ascending=False, method='first') 
    ranking_asc  = ret.rank(axis=1, ascending=True,  method='first')
    top_signal    = (ranking_desc <= Top_n).astype(int)
    bottom_signal = (ranking_asc  <= Low_N).astype(int)*(-1)
    combined_signal = top_signal + bottom_signal

    signal_bool = {"long_short": combined_signal}
    
    filled_signals = {
        k: fill_signal(v, ret)
        for k, v in signal_bool.items()
    }
    # -------------------------------
    # (3) 在 rebalancing 日檢查: 若該股票有活躍可轉債 => 用可轉債
    if (cb_ret is not None) and (cb_adj_close is not None) and (cb_exp_ret is not None):
        for day in filled_signals['long_short'].index:
            if day not in month_end_dates:
                continue

            active_stocks = filled_signals['long_short'].loc[day]
            active_stocks = active_stocks[active_stocks != 0].index  # 只取有訊號的股票
            for stock_ticker in active_stocks:
                cb_col = find_active_cb_for_stock(
                    stock_ticker, day, cb_ret=cb_ret, lookback_days=lookback_days
                )
                print(f'{day}: {stock_ticker} => {cb_col}')
                if cb_col is not None:
                    # 轉移訊號: 原本 stock_ticker => cb_col
                    val = filled_signals['long_short'].at[day, stock_ticker]
                    # 清掉原股票
                    filled_signals['long_short'].at[day, stock_ticker] = 0
                    # 若 cb_col 不在 columns, 先補一欄
                    if cb_col not in filled_signals['long_short'].columns:
                        filled_signals['long_short'].loc[:, cb_col] = 0.0
                    # 設定訊號
                    filled_signals['long_short'].at[day, cb_col] = val

                    # 同時, 需要確保 adj_close, exp_ret 也有 cb_col
                    # 若沒有, 則把 cb_adj_close[cb_col], cb_exp_ret[cb_col] 填到相同 DataFrame
                    if cb_col not in adj_close.columns:
                        adj_close[cb_col] = cb_adj_close[cb_col]
                    if cb_col not in exp_ret.columns:
                        exp_ret[cb_col] = cb_exp_ret[cb_col]

                    if cb_col in adj_close.columns and cb_col in exp_ret.columns:
                        print(f"Before weighting, {cb_col} price stats:\n", adj_close[cb_col].describe())
                        print(f"Before weighting, {cb_col} exp_ret stats:\n", exp_ret[cb_col].describe())
    # -------------------------------
    # (4) 計算權重 (W_df)
    # -------------------------------
    W_df = {
        k: (signal.astype(int) / (Top_n + Low_N)) 
        for k, signal in filled_signals.items()
    }

    def calculate_weighting(W):
        weighting = ((np.floor((W * AUM) / (1000 * adj_close))) * (1000 * adj_close)) / AUM
        weighting.replace({0: np.nan}, inplace=True)
        return weighting

    weightings = {
        k: calculate_weighting(W) 
        for k, W in W_df.items()
    }
    # -------------------------------
    # (5) 停損
    # -------------------------------
    for idx in weightings['long_short'].index:
        daily_ret_series = exp_ret.loc[idx]
        stop_stocks = daily_ret_series[daily_ret_series <= stop_loss_threshold].index
        weightings['long_short'].loc[idx, stop_stocks] = 0.0

    # -------------------------------
    # (6) 計算每日組合報酬
    # -------------------------------
    def calculate_return(weighting):
        delta_weighting = weighting.diff()
        # 買入費用(買=delta>0)
        day_buy_fee  = delta_weighting[delta_weighting > 0] * (0.001425 * 0.3 + 0.003)
        # 賣出費用(賣=delta<0)
        day_sell_fee = delta_weighting[delta_weighting < 0].abs() * (0.001425 * 0.3)
        return ((weighting * exp_ret) - day_buy_fee.fillna(0) - day_sell_fee.fillna(0)).sum(axis=1)

    return_series = {
        k: calculate_return(weighting) 
        for k, weighting in weightings.items()
    }

    for k, v in return_series.items():
        v.iplot(title=k)

    bt_ret = pd.concat({
        'ls_return_cb': return_series['long_short'],
        'benchmark': bench_exp_ret.reindex(exp_ret.index).fillna(0),
    }, axis=1).loc['2020':].iloc[:-2] 

    return bt_ret, weightings

bt_ret1, weightings1 = calculate_return_series(
    ret=pre_ret,
    Top_n=top_n,
    Low_N=low_n,
    AUM=AUM,
    adj_close= Data_dict['Adj_Close'][pool],
    exp_ret=exp_ret[pool],
    cb_ret=cb_ret,
    cb_adj_close=cb_adj_close,
    cb_exp_ret=cb_exp_ret,
    stop_loss_threshold=-0.05,
    lookback_days=5,
    bench_exp_ret=bench_exp_ret
)
```

```python
def max_drawdown(prices):
    # 計算累計的最大值
    cumulative_max = prices.cummax()
    # 計算回撤 (Drawdown)
    drawdown = (prices - cumulative_max) / cumulative_max
    # 計算最大回撤 (MDD)
    mdd = drawdown.min()
    return mdd
```
